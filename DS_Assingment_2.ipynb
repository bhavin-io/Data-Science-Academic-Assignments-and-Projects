# Task 1
Write a Pandas program to create a line plot of the historical stock prices of
Alphabet Inc. between two specific dates. For instance start date 2020-4-1
and end date 2020-09-30.

```python
start_date = "2020-04-01"
end_date = "2020-09-30"
fd = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
plt.title("Historical Stock Prices of Alphabet Inc. Line Plot")
plt.xlabel('Dates Between 2020-04-01 and 2020-09-30')
plt.ylabel('Closing Price $')
plt.plot(fd['Date'],fd['Close'])
plt.xticks([])

plt.show()
```

# Task 2
Write a Pandas program to create a stacked bar plot of opening, closing stock
prices of Alphabet Inc. between two specific dates. For instance start date
“2020-4-1” end date “2020-4-30”

```python
start_date = "2020-04-01"
end_date = "2020-04-30"
fd = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
fd[['Open', 'Close']].plot(kind='bar', stacked=True)
plt.title("Opening and Closing Stock Prices of Alphabet Inc. (April 2020)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(range(len(fd)), fd['Date'])
plt.legend(["Open", "Close"], loc="upper left")

plt.show()
```

# Task 3
Write a Pandas program to create a scatter plot of the trading volume/stock prices
of Alphabet Inc. stock between two specific dates. For instance start date
“2020-4-1” end date “2020-4-30”


```python
start_date = "2020-04-01"
end_date = "2020-04-30"
fd = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
x = fd['Volume']
y = fd['Close']
plt.scatter(x, y)
plt.title("Trading Volume vs Stock Prices(April 2020)")
plt.xlabel("Closing Price")
plt.ylabel("Volume")
plt.show()
```
