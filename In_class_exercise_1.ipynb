# Simple interactive plot using Plotly.

```python
import plotly.graph_objects as px
import pandas as pd
data  = pd.read_csv("tips.csv")

plot=px.Figure(data = [px.Scatter(x=data['day'], y=data['tip'],mode='markers',)])

plot.update_layout(
    updatemenus =[
        dict(
            type="buttons",
            direction = "left",
            buttons  = list([
                dict(
                    args = ["type","scatter"],
                    label = "Scatter plot",
                    method="restyle"
                ),
                dict(
                    args=['type','bar'],
                    label = "Bar Chart",
                    method = 'restyle'
                )
            ]),
        ),
    ]
)
plot.show()
```

# A function to plot a line using the Midpoint Line Drawing Algorithm

```python
import matplotlib.pyplot as plt

def plot_line(start_x, start_y, end_x, end_y):
    dx = end_x - start_x
    dy = end_y - start_y
    d = 2 * dy - dx
    y = start_y

    points = [(start_x, start_y)]

    for x in range(start_x + 1, end_x + 1):
        points.append((x, y))
        if d > 0:
            y += 1
            d += 2 * (dy - dx)
        else:
            d += 2 * dy

    return points

def main():
    start_x = int(input("Enter start x-coordinate: "))
    start_y = int(input("Enter start y-coordinate: "))
    end_x = int(input("Enter end x-coordinate: "))
    end_y = int(input("Enter end y-coordinate: "))

    line_points = plot_line(start_x, start_y, end_x, end_y)

    x_values = [point[0] for point in line_points]
    y_values = [point[1] for point in line_points]

    plt.plot(x_values, y_values)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Line Plot Using Midpoint Algorithm')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
