```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
data = pd.read_csv("./tips.csv")
plt.scatter(data['day'], data['tip'],c=data['size'],s=data["total_bill"])
plt.title("Scatter plot")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.colorbar()
plt.show()
```

```python

plt.plot(data['tip'])
plt.plot(data['size'])

plt.title("Line Plot")

plt.xlabel("Day")
plt.ylabel('Tip')

plt.show()
```

```python
plt.bar(data['day'],data['tip'])
plt.title("Bar Chart")
plt.xlabel('Day')
plt.ylabel('Tips')
plt.show()
```

```python
plt.hist(data['total_bill'])
plt.title("histogram")
plt.show()
```

Matplotlib is a Python plotting library designed for creating various types of graphs, including scatter plots, histograms, bar charts, and line charts. In the realm of Data Science, Matplotlib plays a crucial role in visualizing and representing data effectively. Matplotlib works seamlessly with Pandas, a popular data manipulation library in Python. Pandas' Data Frame can be directly used with Matplotlib, simplifying the process of visualizing data stored in tabular form.
