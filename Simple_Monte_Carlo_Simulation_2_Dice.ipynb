# Monte Carlo simulation of rolling two six-sided dice and visualizing the results using a histogram

```python
import random
import matplotlib.pyplot as plt

def roll_dice():
    return random.randint(1, 6), random.randint(1, 6)

def monte_carlo_simulation(num_trials):
    results = []
    for _ in range(num_trials):
        roll1, roll2 = roll_dice()
        results.append(roll1 + roll2)
    plt.hist(results, bins=range(2, 14), density=True, align='left')
    plt.xlabel('Sum of Two Dice')
    plt.ylabel('Probability')
    plt.title('Monte Carlo Simulation of Rolling Two Six-sided Dice')
    plt.xticks(range(2, 13))
    plt.grid(axis='y', alpha=0.75)
    plt.show()

num_trials = 1000000
monte_carlo_simulation(num_trials)
```
